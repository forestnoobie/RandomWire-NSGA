# Deap implementation

```python
import random

from deap import base
from deap import creator
from deap import tools

creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("FitnessMin", base.Fitness, weights=(0))
creator.create("Individual", list, fitness=creator.FitnessMax)

IND_SIZE =  3

toolbox = base.Toolbox()

## 초기에 만들어진 갯수만큼 뒤에 49를 바꿔준다!
toolbox.register("attr_int", random.randint,0,49)
toolbox.register("individual", tools.initRepeat, creator.Individual,
                toolbox.attr_int, n=IND_SIZE)
```

```python
## population 생성
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

```python
import sys
sys.path.insert(0,'../')
from ga_utils.ga import ga_trainer
import easydict

## 재훈련 방지를 위해 val_acc를 담은 파일
# key : index1.index2.indexx.3
master_dict = {}

args = easydict.EasyDict({
 
        "config": './config/config.yaml',
 
        "checkpoint_path": None,
 
        "model": 'ws-4-0.75'
})

def mutation(individual):
    individual = tools.mutUniformInt(individual,0,50,0.02)
    return individual

def fitness_function(individual_list):
    
    global master_dict
    index_key = '{}.{}.{}'.format(individual_list[0],individual_list[1],individual_list[2])
    
    if index_key in master_dict.keys():
        return (master_dict[index_key],)
    
    else :
        f_path = 'D:/skt/rw_ga/graphs/ga_initials'
        fname = 'ws-4-0.75'
        args = easydict.EasyDict({
            "config": '../config/config.yaml',

            "checkpoint_path": None,

            "model": 'ws-4-0.75'
    })
        ## 위 filename list로 yaml파일 수정하기
        #여기에 ga_trainer 함수 , 파일 리스트 주면 모델 트레이닝
        val_acc = ga_trainer(args, individual_list ,f_path,fname)
        master_dict[index_key] = val_acc
        return (val_acc,)
```

```python
toolbox.register("mate",tools.cxOnePoint)
toolbox.register("select",tools.selRoulette, fit_attr='fitness')
toolbox.register("evaluate", fitness_function)
toolbox.register("mutate",mutation)
```

### Crossover, mutation

```python
## one point Cross over
## mutation change index : mutUniformInt

def mutation(individual):
    individual = tools.mutUniformInt(individual,0,50,0.02)
    return individual
toolbox.register("mutate",mutation)
```

```python
NGEN = 3
start = time.time()

for g in range(NGEN):
    # Select the next generation individuals
    offspring = toolbox.select(pop, len(pop))
    # Clone the selected individuals
    offspring = list(map(toolbox.clone, offspring))

    # Apply crossover on the offspring
    for child1, child2 in zip(offspring[::2], offspring[1::2]):
        if random.random() < 0.5:
            toolbox.mate(child1, child2)
            del child1.fitness.values
            del child2.fitness.values

    # Apply mutation on the offspring
    for mutant in offspring:
        if random.random() < 0.2:
            toolbox.mutate(mutant)
            del mutant.fitness.values

    # Evaluate the individuals with an invalid fitness
    invalid_ind = [ind for ind in offspring if not ind.fitness.valid]
    fitnesses = toolbox.map(toolbox.evaluate, invalid_ind)
    for ind, fit in zip(invalid_ind, fitnesses):
        ind.fitness.values = fit

    # The population is entirely replaced by the offspring
    pop[:] = offspring
elapse = time.time() - start
```
