```python
import json
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
data_path = "C:/Users/82103/Desktop/datascience/rwnn_git/result/"
```

```python
fname = "train_logging.txt"


with open(os.path.join(data_path,fname)) as json_file:
    data = json.load(json_file)

train_log = data['train_log']

objs =[]
npop = 8
niter = len(train_log)
for i in range(niter):
    objs.append(train_log[str(i)])

    
    
## Accuracy

objs = np.array(objs)
idx = 0

epoch = list(range(niter))
obj1 = np.min(objs[:,:,idx],axis=-1)
obj1 = obj1 * - 1
obj1_mean = np.mean(objs[:,:,idx],axis=-1)
obj1_mean = obj1_mean * -1

plt.plot(epoch, obj1, 'g', label = 'Max Accuracy')
plt.plot(epoch, obj1_mean, 'r', label = "Mean Accuracy")
plt.xlabel("Generation")
plt.ylabel("Accuracy")
plt.legend()
plt.savefig("Accuracy_Plot.png", dpi = 100)
plt.show()
```

```python
## Flops
idx = 1

epoch = list(range(niter))
obj1 = np.min(objs[:,:,idx],axis=-1)
obj1 = obj1
obj1_mean = np.mean(objs[:,:,idx],axis=-1)
obj1_mean = obj1_mean

plt.plot(epoch, obj1, 'g', label = 'Max Flops')
plt.plot(epoch, obj1_mean, 'r', label = "Mean Flops")
plt.xlabel("Generation")
plt.ylabel("Flops")
plt.legend()
plt.savefig("Flop_plot.png", dpi = 100)
plt.show()
```

## Plotly Visualization

### 개발해야 할 항목
* json -> Data parsing
* Data -> Plotly graph

```python
# Modularize
import json
import os
import numpy as np
import plotly
import plotly.graph_objects as go

plotly.offline.init_notebook_mode(connected=True)

def json2graph(path,gr_type):
    # path : log file path
    # gr_type 0 : acc , 1: flops
    
    if gr_type == 0 : 
        gr_name ="GA acc"
    else : gr_name = "GA flops"
    
    with open(os.path.join(data_path,"train_logging.log")) as json_file:
        data = json.load(json_file)
        
    train_log = data['train_log']
    niter = len(train_log)
    npop = len(train_log['0'])
    objs =[]

    for i in range(niter):
        temp = [train_log[str(i)][j][1] for j in range(npop)]
        objs.append(temp)

    objs = np.array(objs)
    epoch = list(range(niter))

    # 0 : acc , 1: Flops
    idx = gr_type
    objs = -1*objs
    y1_lower = np.min(objs[:,:,idx],axis=-1)
    y1_lower = list(y1_lower[::-1])

    y1 = np.mean(objs[:,:,idx],axis=-1)

    y1_upper = np.max(objs[:,:,idx],axis=-1)
    y1_upper = list(y1_upper[::-1])


    x = [i+1 for i in range(niter)]
    x_rev = x[::-1]

    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=x+x_rev,
        y=y1_upper+ y1_lower,
        fill='toself',
        fillcolor='rgba(0,100,80,0.2)',
        line_color='rgba(255,255,255,0)',
        showlegend=False,
        name=gr_name,
    ))

    fig.add_trace(go.Scatter(
        x=x, y=y1,
        line_color='rgb(0,100,80)',
        name=gr_name,
    ))

    fig.update_traces(mode='lines')
    fig.show()
```

```python
data_path = "C:/Users/82103/Desktop/datascience/rwnn_git/RandomWire-NSGA/notebooks/logs/new_experiment/"
json2graph(data_path,0)
```
